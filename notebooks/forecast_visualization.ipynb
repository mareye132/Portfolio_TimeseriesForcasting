```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the forecast results from the CSV file
file_path = 'C:/Users/user/Desktop/Github/Portfolio_TimeseriesForcasting/notebooks/data/forecast_results.csv'  # Ensure this path is correct relative to the notebook location
forecast_data = pd.read_csv(file_path, parse_dates=['Date'])

# Check the first few rows of the dataset to verify the data
forecast_data.head()

# Plot the actual vs. forecasted values
plt.figure(figsize=(14, 7))
plt.plot(forecast_data['Date'], forecast_data['Close'], label='Actual', color='blue')
plt.plot(forecast_data['Date'], forecast_data['Forecast'], label='Forecasted', color='red')
plt.title('Actual vs. Forecasted Stock Price')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.legend()
plt.grid(True)
plt.show()

# Calculate and display the evaluation metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error

# Assuming 'Close' is the actual price and 'Forecast' is the predicted value
mae = mean_absolute_error(forecast_data['Close'], forecast_data['Forecast'])
rmse = mean_squared_error(forecast_data['Close'], forecast_data['Forecast'], squared=False)
mape = mean_absolute_percentage_error(forecast_data['Close'], forecast_data['Forecast'])

print("Evaluation Metrics:")
print(f"MAE (Mean Absolute Error): {mae:.2f}")
print(f"RMSE (Root Mean Squared Error): {rmse:.2f}")
print(f"MAPE (Mean Absolute Percentage Error): {mape*100:.2f}%")
```
